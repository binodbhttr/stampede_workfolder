```python
import utilities as ut
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sl_utilities import distinct_colours as dc
from sl_utilities import distance_functions
import os
import pickle
import imageio

path="./data_pkl/" 
snap=596
simname = 'm12i_res7100_mhdcv'
simdir = '/scratch/projects/xsede/GalaxiesOnFIRE/mhdcv/m12i_res7100_mhdcv/1Myr/1Myr_fire2'
file_name="total_data_all_clusters_all_snapshots.pkl"
snapshot_start=596
snapshot_end=696



with open(path+file_name, "rb") as input:
    importdata = pickle.load(input)

print("Testing if the loading of data was successful !! \n")
print("x_tracked of snapshot 596 for cluster 1",importdata[596][1]["x_tracked"])
print("x_cm of snapshot 596 for cluster 1",importdata[596][1]["xcm"])
print("x_tracked of snapshot 597 for cluster 1",importdata[597][1]["x_tracked"])
print("x_cm of snapshot 597 for cluster 1",importdata[597][1]["xcm"])
print("x_tracked of snapshot 597 for cluster 2",importdata[597][2]["x_tracked"])
print("x_cm of snapshot 597 for cluster 2",importdata[597][2]["xcm"])



cluster_groupid=[]
total_clusters=len(importdata[snapshot_start])
for i in range(total_clusters):
    cluster_groupid.append("snapshot"+str(snapshot_start)+"_cluster_group"+str(i+1))

print("These are the clusters groups we have tracked data of:\n",cluster_groupid)
```

```python
#Here we track a star cluster keeping the axes limits fixed
#we calculated the maximum and minimum possible values for x and y axes and set the limits accordingly
plot_path="./plots_pkl/animation_plots/" #creating a path to store the plots only if it does not exist
if not os.path.exists(plot_path):
  os.makedirs(plot_path)


n=snapshot_end-snapshot_start+1
colors=['cyan','blue','green','magenta','yellow','teal','brown','tan','lime','red']

total_clusters=1
#fig1.suptitle("Cylindrical component of the velocity of Cluster Groups from Snapshot 596")
plot_count=0
#plot_count=0
snapshot_list=np.arange(snapshot_start,snapshot_end+1) #create a list of snapshots to plot
time=snapshot_list-snapshot_start #time starts from zero here where t=0 is at the snapshot where we start from
#time=[[0],[20],[40],[60],[80],[100]]

xmax=np.amax(importdata[time[1]+snapshot_start][1]["x_tracked"])
xmin=np.amin(importdata[time[1]+snapshot_start][1]["x_tracked"])
ymax=np.amax(importdata[time[1]+snapshot_start][1]["y_tracked"])
ymin=np.amin(importdata[time[1]+snapshot_start][1]["y_tracked"])
for i in range(len(time)-10):
    cluster_count=0
    
    for j in range(total_clusters):
        x=importdata[time[i]+snapshot_start][cluster_count+1]["x_tracked"]
        xmax_new=np.amax(x)
        print(xmax_new)
        xmin_new=np.amin(x)
        if xmax_new>xmax:
            xmax=xmax_new
        if xmin_new<xmin:
            xmin=xmin_new
        
        y=importdata[time[i]+snapshot_start][cluster_count+1]["y_tracked"]
        ymax_new=np.amax(y)
        ymin_new=np.amin(y)
        if ymax_new>ymax:
            ymax=ymax_new
        if ymin_new<ymin:
            ymin=ymin_new

print(xmax,xmin,"ymaxmin",ymax,ymin)
for i in range(len(time)):
    
    fig2=plt.figure(figsize=(8,5))
    ax = fig2.add_subplot(1,1,1)
    cluster_count=0
    for j in range(total_clusters):
        
        x=importdata[time[i]+snapshot_start][cluster_count+1]["x_tracked"]
        y=importdata[time[i]+snapshot_start][cluster_count+1]["y_tracked"]
        xcm=importdata[time[i]+snapshot_start][cluster_count+1]["xcm"]
        ycm=importdata[time[i]+snapshot_start][cluster_count+1]["ycm"]
        z=importdata[time[i]+snapshot_start][cluster_count+1]["z_tracked"]
        
        ax.scatter(x,y,label="cluster_group"+str(cluster_count+1),c=x)
        ax.scatter(xcm,ycm,c="black")
        ax.minorticks_on()
        ax.set_xlabel("x position in kpc")
        ax.set_ylabel("y position in kpc")
        ax.set_title("Clusters at T="+str(time[plot_count])+" in Myr")
        ax.set_xlim(xmin-abs(xmin/10),xmax+abs(xmax/10))
        ax.set_ylim(ymin-abs(ymin/10),ymax+abs(ymax/10))
        cluster_count+=1
    
    ax.legend(bbox_to_anchor=(1,0.5), loc='center left')
    plt.tight_layout()
    fig2.savefig(plot_path+"cluster_1_fixed_axes_snap"+str(plot_count)+".png",bbox_inches='tight',dpi=200)
    plot_count+=1
    fig2.clf()
    plt.close()
        
#fig1.subplots_adjust(right=1.5)    
#fig11.legend((s1), ("vphi_cyl"),"lower right")
#fig11.suptitle("This is the title \n")

cluster1_snaps=[]
for k in range(0,101):
    cluster1_snaps.append(imageio.imread(plot_path+"cluster_1_fixed_axes_snap"+str(k) + '.png'))
    os.remove(plot_path+"cluster_1_fixed_axes_snap"+str(k) + ".png")
imageio.mimsave(plot_path+"cluster1_fixed_axes.gif", cluster1_snaps, duration = 1/8)
```

```python
#Here we try to track a star cluster keeping the center of mass at the center of the plot
plot_path="./plots_pkl/animation_plots/" #creating a path to store the plots only if it does not exist
if not os.path.exists(plot_path):
  os.makedirs(plot_path)


n=snapshot_end-snapshot_start+1
colors=['cyan','blue','green','magenta','yellow','teal','brown','tan','lime','red']

total_clusters=1
#fig1.suptitle("Cylindrical component of the velocity of Cluster Groups from Snapshot 596")
plot_count=0
#plot_count=0
snapshot_list=np.arange(snapshot_start,snapshot_end+1) #create a list of snapshots to plot
time=snapshot_list-snapshot_start #time starts from zero here where t=0 is at the snapshot where we start from
#time=[[0],[20],[40],[60],[80],[100]]

#################################
################################
rmax_global=importdata[time[1]+snapshot_start][1]["rmax"]
for i in range(len(time)):
    cluster_count=1 #change it to look for cluster no cluster_count+1
    
    for j in range(total_clusters):
        r=importdata[time[i]+snapshot_start][cluster_count+1]["rmax"]
        
        if r>rmax_global:
            rmax_global=r


#######################################
######################################

for i in range(len(time)):
    
    fig2=plt.figure(figsize=(7,5))
    ax = fig2.add_subplot(1,1,1)
    cluster_count=1 #change it change it to look for cluster no cluster_count+1
    for j in range(total_clusters):
        
        x=importdata[time[i]+snapshot_start][cluster_count+1]["x_tracked"]
        y=importdata[time[i]+snapshot_start][cluster_count+1]["y_tracked"]
        z=importdata[time[i]+snapshot_start][cluster_count+1]["z_tracked"]
        xcm=importdata[time[i]+snapshot_start][cluster_count+1]["xcm"]
        ycm=importdata[time[i]+snapshot_start][cluster_count+1]["ycm"]
        zcm=importdata[time[i]+snapshot_start][cluster_count+1]["zcm"]
        rmax=importdata[time[i]+snapshot_start][cluster_count+1]["rmax"]
        xlim_max=xcm+1.1*rmax_global
        xlim_min=xcm-1.1*rmax_global
        
        zlim_max=zcm+1.1*rmax_global
        zlim_min=zcm-1.1*rmax_global
        
        ylim_max=ycm+1.1*rmax_global
        ylim_min=ycm-1.1*rmax_global
        #print("These are the dimensions of the plot",xlim_max-xlim_min,ylim_max-ylim_min)
        ax.scatter(x,y,label="cluster_group"+str(cluster_count+1),c=x)
        ax.scatter(xcm,ycm,c="black",marker='x')
        ax.minorticks_on()
        ax.set_xlabel("X")
        ax.set_ylabel("Y")
        
        ax.set_title("Cluster "+str(cluster_count+1)+" at T="+str(time[plot_count])+" in Myr")
        ax.set_xlim(xlim_min,xlim_max)
        ax.set_ylim(ylim_min,ylim_max)
        ax.set_xticks([], [])
        ax.set_yticks([], [])
        #ax.axis('off') This removes everything including the box and labels.
        ax.set_aspect('equal')
        ax.annotate('', xy=(xcm-0.5, ylim_min+0.2), xytext=(xcm+0.5, ylim_min+0.2),xycoords='data', arrowprops=dict(alpha=0.5, fc='b', ec='b', arrowstyle="<->"))
        ax.annotate('1 kpc',xy=(xcm,ylim_min+0.3),textcoords='data')
        #ax.arrow(xlim_min,ylim_min, xlim_min+0.9, 0,head_width=0.05, head_length=0.1, fc='k', ec='k')
        cluster_count+=1
    
    #ax.legend(bbox_to_anchor=(1,0.5), loc='center left')
    plt.tight_layout()
    #plt.axis('square')
    fig2.savefig(plot_path+"cluster_2_snap_fixedCM"+str(plot_count)+".png",bbox_inches='tight',dpi=200)
    plot_count+=1
    fig2.clf()
    plt.close()
        
#fig1.subplots_adjust(right=1.5)    
#fig11.legend((s1), ("vphi_cyl"),"lower right")
#fig11.suptitle("This is the title \n")

cluster1_snaps=[]
for k in range(0,101):
    cluster1_snaps.append(imageio.imread(plot_path+"cluster_2_snap_fixedCM"+str(k) + '.png'))
    os.remove(plot_path+"cluster_2_snap_fixedCM"+str(k) + ".png")

imageio.mimsave(plot_path+"cluster2_fixedCM.gif", cluster1_snaps, duration = 1/8)
```
